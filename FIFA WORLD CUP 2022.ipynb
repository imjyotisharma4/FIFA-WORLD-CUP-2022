```python
# needs both rows and columns on index subsetting (R,C format)
# index based subsetting(iloc)
df1.iloc[2:5,3:5]
```

```python
df2=df1.iloc[:,1:6]
df2
```

```python
df2=df1.loc[:,["Source of Tweet","Number of Likes","Tweet","Sentiment"]]                 #value based subsetting
df2
```

```python
df3=df2.rename(columns={"Sentiment":"Response","Number of Likes":'Number of Likes each day'})                #sentiment renamed as Response
df3
```

```python
df3_copy=df3.copy()                           
df3_copy
```

```python
mask=df3_copy["No. of likes per week"]!=0                                  #not equal to '0'
df3_copy[mask]
```
